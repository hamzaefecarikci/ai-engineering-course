```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ["connect", "connected", "connecting", "connection", "connections", "connects"]
```

```python
for t in connect_tokens:
    print(t,"\t" ": ", ps.stem(t))
```

```python
learn_tokens = ["learn", "learned", "learning", "learner", "learns", "learner","learners"]
```

```python
for t in learn_tokens:
    print(t, ": ", ps.stem(t))
```

### Lemmetization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for t in connect_tokens:
    print(t, ": ", lemmatizer.lemmatize(t))
```

```python
for t in learn_tokens:
    print(t, ": ", lemmatizer.lemmatize(t))
```
